```python
import zipfile
import os

# Set your file and output folder paths
zip_path = '/content/drive/MyDrive/additional-resources/lada000b21a_e.zip'
extract_dir = '/content/drive/MyDrive/additional-resources/ADA_shapefile_unzipped/'  # Folder to save extracted files

# Create output directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print("✅ Shapefile unzipped successfully.")
```

```python

import geopandas as gpd

# Find the .shp file in the extracted folder
for file in os.listdir(extract_dir):
    if file.endswith('.shp'):
        shp_file_path = os.path.join(extract_dir, file)
        break

# Load shapefile
ada_shapes = gpd.read_file(shp_file_path)
ada_shapes.head()
```
