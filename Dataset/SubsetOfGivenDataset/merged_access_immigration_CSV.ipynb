# **merged both given dataset on column ADA_code **

```python
import pandas as pd

# Load both datasets
access_df = pd.read_csv('/content/drive/MyDrive/Dataset/ADA_acs_file.csv')
immigration_df = pd.read_csv('/content/drive/MyDrive/Dataset/ADA_profile_simplified.csv')

# Merge on ADA_code
merged_df = pd.merge(access_df, immigration_df[['ADA_code', 'PRNAME', 'T1528', 'T1536', 'T1537']], on='ADA_code', how='left')

# Save to a new file to use in Excel
merged_df.to_csv('/content/drive/MyDrive/Dataset/merged_access_immigration2.csv', index=False)

print("Merged file saved")
```

```python
import pandas as pd

# Load local CSVs
access_df = pd.read_csv('/content/drive/MyDrive/Dataset/ADA_acs_file.csv')
immigration_df = pd.read_csv('/content/drive/MyDrive/Dataset/ADA_profile_simplified.csv')

# Keep relevant columns only for RQ4
relevant_columns = ['ADA_code', 'T1529', 'W1529', 'T1528', 'T1536']
immigration_df_subset = immigration_df[relevant_columns]

# Merge on ADA_code
merged_df = pd.merge(access_df, immigration_df_subset, on='ADA_code', how='left')

# Compute number of immigrant men
merged_df["M1529"] = merged_df["T1529"] - merged_df["W1529"]

# Save the merged file
merged_df.to_csv("merged_access_employment_RQ4.csv", index=False)

print("Merged file saved as 'merged_access_employment_RQ4.csv'")
```
